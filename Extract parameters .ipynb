```python
def get_possible_params(function):
    values_to_replace = (('optional', ''),('boolean', ''),('bool', ''),('int', ''),
                        ('array', ''),('or', ''),(' and ', ''),(',', ''),
                        ('str', ''),('float', ''),('dict', ''),(')', ''),('(', ''),
                        ('None', ''),(':', ''),('=', ''),(')', ''),('(', ''))

    values_to_replace2 = (('='       , ''),(')'       , ''),('('       , ''))
    
    website_url = requests.get('https://scikit-learn.org/stable/modules/classes.html').text
    links = BeautifulSoup(website_url,'lxml').find('table')

    i =0
    names = []
    hrefs = []
    while True:
        try:
            links = links.find_next('a')
            names.append(links.text)
            hrefs.append("https://scikit-learn.org/stable/modules/" +links.get('href').split('#',1)[0])
        except:
            break

    df_complete = pd.DataFrame([names, hrefs]  ).T
    df_complete.columns = ['Function', 'link']
    df_complete.set_index('Function', inplace = True)
    df_complete=df_complete[df_complete.link != 'https://scikit-learn.org/stable/modules/']

    website_url = requests.get(df_complete.loc[function].values[0]).text
    links = BeautifulSoup(website_url,'lxml').find('table').find('td').find_next('dt')
    dict_info = {}

    while links.text.split(' ', 1)[1:] != [] and links.text.split(' ', 1)[0][0] != '\n':
        words = links.text.split(' ', 2)[2:][0]
        
        try:
            values = words.split('default',1)[0]
            for r in values_to_replace:
                values = values.replace(*r)
                
            default =  words.split('default',1)[1]
            for r in values_to_replace2:
                default = default.replace(*r)  
                
            dict_info[links.text.split(' ', 1)[0]]=[
                str(values),
                str(default),
                str(words.__contains__('optional')),
                str(words.__contains__('bool' )),
                str(words.__contains__('int'  )),
                str(words.__contains__('None' )),
                str(words.__contains__('array')),
                str(words.__contains__('str'  )),
                str(words.__contains__('float')),
                str(words.__contains__('dict' ))
            ]
            links = links.find_next('dt')
        except:
            break

    df_final = pd.DataFrame(dict_info, ).T
    df_final.columns = ['values','default', 'optional', 'support_bool', 'support_int', 'support_None', 'support_array', 'support_str', 'support_float', 'support_dict']

    return(df_final)
```
